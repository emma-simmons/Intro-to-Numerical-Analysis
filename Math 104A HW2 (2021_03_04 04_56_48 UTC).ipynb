$\textbf{Question 3(a):}$

```python
import math 
import numpy as np
from matplotlib import pyplot as plt 

def l_j(x, j, x_k): #implements the l_j in the Lebesgue function, x is what it's evaluated at, j is to tell if i=j, x_k are the nodes
    p= 1 
    for i in range(0, len(x_k)):
        if i!=j:
            p= p*((x-x_k[i])/(x_k[j]-x_k[i]))
    return p #returns the value of l_j

def L_n(x, x_k): #implements Lebesgue function, x is what it's evaluated at, x_k are the nodes
    L_sum=0
    for j in range(0, len(x_k)):
        L_sum+=np.abs(l_j(x, j, x_k))
    return L_sum #returns the value of the Lebesgue function at x
    
```

$\textbf{Question 4 Part 1:}$

```python
def lambda_j(x_j, n): #computes the barycentric weights, x_j are nodes, n is n in x_n, our nodes
    l=np.ones(n+1)
    for m in range(1,n+1):
        for j in range(m):
            l[j]=l[j]/(x_j[j]-x_j[m])
        L=1
        for i in range(m):
            L*=(1/(x_j[m]-x_j[i]))
        l[m]=L
    return l #returns lamda_j for x_j nodes
                
def barycentric(x, x_j, f_xj, n): #implements barycentric formula, x is what it's being evaluated at, x_j is nodes, f_xj is our nodes plugged into f, n is n in x_n
    top=0
    btm=0
    for j in range(0,n+1):
        B_sum= lambda_j(x_j, n)[j]/(x-x_j[j])
        top+= B_sum*f_xj[j]
        btm+= B_sum
    return top/btm #returns value of barycentric formula at a value of x  
```
